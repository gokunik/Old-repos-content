```python
# Q2: Write a menu driven program in C to create a structure employee having fields empid, empname, salary. 
# Accept the details of 'n' Employees from user and perform the following operations using function. 

# Search by Emp-Id
# Display All
# Display Names of Employee having Salary > 10000.

def display(emp_id,emp_name,emp_salary,emp):
    
    print("Employee id:     ",emp_id)
    print("Employee Name:   ",emp_name)
    print("Employee Salary: ",emp_salary)
    print("\n")
    

name = input("Enter your Name : ")

emp=[]
choice=0
stopper=0

print("\n")
print("welcome ",name)
print("\n")

while stopper==0:
    
    print("press 1 - Enter Employee details")
    print("Press 2 - Search Employees By id.")
    print("Press 3 - Display All Employees.")
    print("Press 4 - Display Names of Employee having Salary > 10000.")
    print("press 5 - Exit")
    choice=int(input())
    
    if choice==1:
        no=int(input("Enter the number of Employees: "))
        for i in range(0,no,1):
            emp_id=input("Enter Emp_id : ")
            emp_name= input("Enter Empolyee Name : ")
            emp_salary=int(input("Enter Salary"))
            print("\n")
            details=(emp_id,emp_name,emp_salary)
            emp.append(details)
            
    elif choice==2:
        search=input("Enter Emp Id: ")
        
        for emp_id,emp_name,emp_salary in emp:
            if emp_id==search:
                display(emp_id,emp_name,emp_salary,emp)
                break
                
    elif choice==3:
        for emp_id,emp_name,emp_salary in emp:
            display(emp_id,emp_name,emp_salary,emp)
               
                
    elif choice==4:        
        for emp_id,emp_name,emp_salary in emp:
            if emp_salary > 10000:
                display(emp_id,emp_name,emp_salary,emp)
        
    elif choice==5:
        print("Program Exited")
        stopper=1
        
```
