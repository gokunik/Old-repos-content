## Question 2) Write a Python program to calculate % of survivors, recovered from Covid-19 in USA with the help of function. % of survivors must be returned by the function. All required values must be entered on run time like total cases, no. of survivors etc.

```python
# python program to find the % of survivors
# https://www.worldometers.info/coronavirus/ and google
# Total Number Of Covid Cases In USA  :- 10063671
# Total Number of Active Cases In USA :- 3429080
# total Number Of Recovered Cases In USA :- 6392330

def survivors(to,ac,re):
    
    stats = {'death_rate':0,'recovery_rate':0,'deaths':0}
    stats['deaths'] = to - ac - re
    stats['recovery_rate'] = re/to * 100
    stats['death_rate'] = stats['deaths']/to * 100
    
    return stats
    
    

total_case  =  int(input('Enter Total Number Of Covid Cases In USA  :- '))
active_case =  int(input('Enter Total Number of Active Cases In USA :- '))
recovered_case =  int(input('Enter Total Number Of Recovered Cases In USA :- '))

stats = survivors(total_case,active_case,recovered_case)

print('\n************ Statistics ************')
print('Total Case      :- ',total_case)
print('Active Case     :- ',active_case)
print('Total Deaths    :- ',stats['deaths'])
print('Death Rate      :- ',round(stats['death_rate'],3),'%')
print('Total Recovery  :- ',recovered_case)
print('Recovery Rate   :- ',round(stats['recovery_rate'],3),'%')
```
