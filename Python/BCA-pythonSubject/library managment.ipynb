```python
# Q1: Write a menu driven program in python which shows the working of library. The menu option should be 
# i) Add book details.
# ii) Display book details.
# iii) List all books of given author.
# iv) List the count of books in the library.
# v) Exit.



name = input("Enter Name : ")

books=[]
choice=0
stopper=0
while stopper==0:
    print("\n")
    print("welcome ",name)
    print("\n")
    print("****************************")
    print("* Library Managment System *")
    print("****************************")

    print("Press 1 - Add book details.")
    print("Press 2 - Display book details.")
    print("Press 3 - List all books of given author.")
    print("Press 4 - List the count of books in the library.")
    print("press 5 - Exit")
    choice=int(input())

    if choice==1:
        no=int(input("Enter the number of books you want to add"))
        for i in range(0,no,1):
            book_no=input("Enter Book Number : ")
            book_name= input("Enter Book Name : ")
            book_author=input("Enter Book Author")
            book_price=int(input("Enter Book Price"))
            print("\n")
            details=(book_no,book_name,book_author,book_price)
            books.append(details)

    elif choice==2:
        print("Press 1 for Displaying Details Of particular book")
        for book_no,book_name,book_author,book_price in books:
            print("Book Number: ",book_no)
            print("Book Name: ",book_name)
            print("Book author: ",book_author)
            print("Book price: ",book_price)
            print("\n")
    
    elif choice==3:
        author=input("Enter Author Name")
        
        for book_no,book_name,book_author,book_price in books:
            if book_author==author:
                print("Book Number: ",book_no)
                print("Book Name: ",book_name)
                print("Book author: ",book_author)
                print("Book price: ",book_price)
                print("\n")
        
    elif choice==4:
        print("Total Number Of Books In library : ",len(books))
        
    elif choice==5:
        print("Program Exited")
        stopper=1
                
            
            


    
```
