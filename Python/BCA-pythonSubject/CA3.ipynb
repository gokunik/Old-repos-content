## Question 1.  Write a Python program to take input of 10 numbers, find the square of all 10 numbers and checkwhether the square of every number is palindrome or not. Numbers where square is palindrome, save infile with original numbers and their squares. Numbers where square is not palindrome, must be saved in separate file. 

```python
# text file is also available in the Github Repo. 
# For file Output Please check file1.txt and file2.txt available in the repo
numbers = []
no = 0
square = 0
pal = ""

try:
    file1 = open('file1.txt','w')
    file1.write('File Containing square of number who\'s square is palindrome \nNumber  |  Square  | Square palindrome ?\n')

    file2 = open('file2.txt','w')
    file2.write('File Containing square of number who\'s square is not palindrome \nNumber  |  Square  | Square palindrome ?\n')

    for i in range(0,10):
        no = int(input('Enter number %d -> '%(i+1)))
        square = no * no
        temp = square
        rev = 0
        while temp > 0:
            rev = (rev * 10) + (temp % 10)
            temp = temp // 10

        if rev == square :
            pal = "Yes"
        else:
            pal = "No"

        numbers.append([no,square,pal])


    for i in range(0,10):
        if(numbers[i][2] == 'Yes'):
            file1.write('%d            %d        %s \n' %(numbers[i][0],numbers[i][1],numbers[i][2]))
        else:
            file2.write('%d            %d        %s \n' %(numbers[i][0],numbers[i][1],numbers[i][2]))


    file1.close()
    file2.close()
    
except Execption as a:
    print('Some Error Occurred -> ',a)
else:
    print(' Program execution finished! data is written to the file ')
    
```

## Question 2. Write a Python program to calculate % of death rate, happened due to Covid-19 in all over INDIA with the help of function. % of death rate must be returned by the function. All required values must beentered on run time like total cases, no. of deaths etc

```python
# data Collected From ->  https://www.worldometers.info/coronavirus/
# on the day when stats are taken statistics numbers are Following ->
# Total Cases    -> 8,464,032
# Total Deaths   -> 125,630
# Recovery cases -> 7,819,886

def calculate(cases,deaths,recovery):
    data = [] 
    death_rate        =  (deaths/cases)       * 100
    recovery_rate     =  (recovery/cases)     * 100
    active_cases      =  (cases - deaths)     - recovery
    death_estimate    =  (death_rate /100)    * 100 # calculating average ratio of death per 100 people
    recovery_estimate =  (recovery_rate /100) * 100 # calculating average ratio of recovery per 100 people
    
    data.append(round(death_rate,2))
    data.append(round(recovery_rate,2))
    data.append(active_cases)
    data.append(round(death_estimate,2))
    data.append(round(recovery_estimate,2))
    
    return data # function is returning data list 
    

total_case =     int(input('Enter total number of cases            -> '))
total_death =    int(input('Enter total number of death cases      -> '))
total_recovery = int(input('Enter total number of recovered cases  -> '))

data = calculate(total_case,total_death,total_recovery)

print('\n')
print('Active cases   -> ',data[2])
print('Death rate      -> ',data[0],' %. At average ',round(data[3]),' out of 100 people died from virus in india ' )
print('Recovery rate  -> ',data[1],' %. At average ',round(data[4]),' out of 100 people Recovered from virus in india ' )


```
